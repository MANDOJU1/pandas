### 데이터 전처리


```python
# , 제거

data = {
    "name":["홍길동","김수정","이민정"],
    "age":["25","35","45"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df["age1"] = df["age"].astype("int64")
df.info()
```

```python
df.loc[1,"age"] = np.nan
df.info()
```

```python
# 타입 변경
# astype("변경할타입") : 데이터에 NaN이 존재한다면 int64 형 변경 불가
# df["age2"] = df["age"].astype("int64") → 오류
df["age2"] = df["age"].astype("float64")

# pd.to_numeric() : 데이터에 NaN 존재 시 float64 로 변경함
df["age3"] = pd.to_numeric(df['age'])
```

```python
# df_man['30~34세'].astype("int64") → 오류 (, 가 있기 때문)
# pd.to_numeric(df_man['30~34세']) → 오류

# , 제거 후 타입변환 필요
df_man['30~34세'].str.replace(",","").astype("int64")
df_man['0~4세'].str.replace(",","").astype("int64")
df_man['20~24세'].str.replace(",","").astype("int64")

# 위에처럼 일일이 만들기는 쉽지않음 → 한번에 타입변경 하는 방법
# apply(함수), map(), applymap()

def comma_split(x):
  return x.str.replace(",","")

# df_man.apply(df_man, axis=0)

# 한번만 타입변경 할거면 함수를 만들필요없이 바로 적용
df_man = df_man.apply(lambda x:x.str.replace(",","").astype("int64"))
df_woman = df_woman.apply(lambda x:x.str.replace(",","").astype("int64"))
```

```python
# 인덱스 공백 : strip()

df_man.index = df_man.index.str.strip()
df_woman.index = df_woman.index.str.strip()
```
