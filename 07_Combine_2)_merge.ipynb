```python
import pandas as pd
```

```python
pd.merge?
```

```python
df1 = pd.DataFrame({"x1":['A','B','C'], "x2":[1,2,3]})
df1
```

```python
df2 = pd.DataFrame({"x1":['A','B','D'], "x2":['T','F','T']})
df2
```

```python
pd.merge(df1, df2, how='left', on='x1')
```

```python
pd.merge(df1, df2, how='right', on='x1')
```

```python
pd.merge(df1, df2, how='inner', on='x1')
```

```python
pd.merge(df1, df2, how='outer', on='x1')
```

* Filterling Join

```python
df2.x1
```

```python
df1.x1
```

```python
# df1.x1 안의 값과 df2.x1 값이 일치한 경우만 가져오기(df1만 내용만)
# df1[조건]

# isin()
# df1[df1.x1.isin(df2.x1)]
df1[df1.x1 == df2.x1]
```

```python
# ~ : 제외하고
df1[~df1.x1.isin(df2.x1)]
df1[df1.x1 != df2.x1]
```

* 두 번째 실습 : indicator 옵션 활용하기 + query() + drop()

```python
df3 = pd.DataFrame({"x1":['B','C','D'], "x2":[2,3,4]})
df3
```

```python
pd.merge(df1, df3)
```

```python
pd.merge(df1, df3, how='outer')
```

```python
pd.merge(df1, df3, how='outer', indicator=True)
```

```python
pd.merge(df1, df3, how='outer', indicator=True).query('_merge == "left_only"')
```

```python
pd.merge(df1, df3, how='outer', indicator=True) \
              .query('_merge == "left_only"') \
              .drop(columns=['_merge'])
```

### [실습2]

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "customer_id": [1, 2, 3],
    "customer_name": [
        "Robert",
        "Peter",
        "Dave",
    ],
})
df1
```

```python
df2 = pd.DataFrame({
    "id": [1, 2, 4],
    "order_id": [100, 200, 300],
    "order_date": [
        "2021-01-21",
        "2021-02-03",
        "2020-10-01",
    ],
})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

#### 인덱스가 있다면?

```python
# 데이터프레임 안의 특정 컬럼을 인덱스로 설정 : set_index(컬럼명)

df1.set_index("id", inplace=True)
```

```python
df2.set_index("id", inplace=True)
```

```python
pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
pd.merge(df1, df2, left_index=True, right_index=True, how='right')
```
